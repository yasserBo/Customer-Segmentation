```python
import pandas as pd

# Load the dataset
data = pd.read_csv('data/Mall_Customers.csv')

# Display the first few rows of the dataframe
print(data.head())

# Get a concise summary of the DataFrame
print(data.info())

# Get a summary of the dataframe
print(data.describe())

# Check for missing values
print(data.isnull().sum())
```

```python
data['Age'] = data['Age'].fillna(data['Age'].median())
```

```python
data = pd.get_dummies(data, columns=['Genre'])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']] = scaler.fit_transform(data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']])
```

```python
import matplotlib.pyplot as plt
data.hist(bins=15, figsize=(15, 6), layout=(2, 3))
plt.show()
```

```python
import seaborn as sns
sns.pairplot(data)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, fmt=".2f")
plt.show()
```

```python
import os
os.environ["OMP_NUM_THREADS"] = "1"  # Set this before importing KMeans or any scikit-learn module
```

```python
from sklearn.cluster import KMeans
# Assuming we decide on 5 clusters
kmeans = KMeans(n_clusters=5, random_state=42)
data['Cluster'] = kmeans.fit_predict(data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']])
```
