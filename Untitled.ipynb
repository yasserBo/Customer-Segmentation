```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data = pd.read_csv('data/Mall_Customers.csv')
print(data.head())
```

```python
# Fill missing values for numeric columns with their mean
for column in data.select_dtypes(include=['float64', 'int64']).columns:
    data[column].fillna(data[column].mean(), inplace=True)

# If you have categorical data with missing values, consider filling with the mode (most frequent category)
for column in data.select_dtypes(include=['object', 'category']).columns:
    data[column].fillna(data[column].mode()[0], inplace=True)
```

```python
# Converting only categorical columns to dummy variables
data = pd.get_dummies(data, columns=data.select_dtypes(include=['object', 'category']).columns, drop_first=True)
```

```python
# Selecting numerical columns to scale
numerical_columns = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']  # adjust as per your dataset

scaler = StandardScaler()
data[numerical_columns] = scaler.fit_transform(data[numerical_columns])
```

```python
inertia = []
for k in range(1, 11):  # Adjust the range as needed
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data[numerical_columns])
    inertia.append(kmeans.inertia_)

# Plotting the elbow curve
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), inertia, marker='o')  # Adjust range accordingly
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
# Using the optimal number of clusters from the elbow method
optimal_clusters = 5  # Replace with your chosen number
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
clusters = kmeans.fit_predict(data[numerical_columns])

# Add cluster information to the DataFrame
data['Cluster'] = clusters
```

```python
# Analyze clusters
cluster_summary = data.groupby('Cluster').mean()
print(cluster_summary)

# Visualization of clusters
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster', data=data, palette='viridis')
plt.title('Clusters of Customers')
plt.show()
```

```python
from sklearn.cluster import KMeans

# Assuming you have a DataFrame named 'data'
# Ensure that the DataFrame 'data' has the columns 'Age', 'Annual Income (k$)', and 'Spending Score (1-100)'

# Set up the KMeans clustering
kmeans = KMeans(n_clusters=5, random_state=42)

# Fit the model and predict clusters
data['Cluster'] = kmeans.fit_predict(data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']])
```

```python
print(data['Cluster'].value_counts())
```
